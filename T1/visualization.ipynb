```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('titanic.csv')
print(df.head())
```

```python
print(df.info())
```

```python
df=df.drop(columns=["PassengerId","Pclass","Name","SibSp","Parch","Ticket","Fare","Cabin","Embarked"])
print(df.head())
```

```python
df.shape
```

```python
plt.hist(df['Age'], bins=20,color='green',edgecolor="black")
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of Titanic Passengers')
plt.savefig("age_histogram.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.bar(df["Sex"].unique(),df["Sex"].value_counts(),color=['blue','green'],edgecolor="black")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.title("Distribution of Sex among Titanic Passengers")
plt.savefig("sex_distribution.png", dpi=300, bbox_inches='tight')
plt.show()
```
